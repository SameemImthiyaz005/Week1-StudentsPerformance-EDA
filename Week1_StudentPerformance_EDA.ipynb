```python
import pandas as pd  # library for handling data
df = pd.read_csv("StudentsPerformance.csv.zip")  # loading the dataset
df.head()  # showing the first 5 rows
```

```python
# Get general info
df.info()
```

```python
# Describe the data (mean, min, max, etc.)
df.describe()
```

```python
df.isnull().sum()  # Checks for missing data
```

```python
# Gender-wise score
df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()
```

```python
# Lunch-wise score
df.groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
```

```python
# Test preparation course effect
df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Gender vs Math Score
sns.boxplot(x='gender', y='math score', data=df)
plt.title("Math Scores by Gender")
plt.show()
```

```python
# Lunch type vs Reading Score
sns.barplot(x='lunch', y='reading score', data=df)
plt.title("Reading Scores by Lunch Type")
plt.show()
```

```python
# Test prep vs Writing Score
sns.boxplot(x='test preparation course', y='writing score', data=df)
plt.title("Writing Scores vs Test Preparation")
plt.show()
```

## Key Insights

1. Females scored higher in reading and writing.
2. Students who completed the test prep course performed better.
3. Standard lunch students scored better overall.
